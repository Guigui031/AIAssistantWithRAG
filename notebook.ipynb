```python
import pandas as pd
```

```python
file_path = 'source-documents/celebrity-cruises.csv'
source_data = pd.read_csv(file_path)
source_data.head()
```

```python
source_data['num_date'] = pd.to_datetime(source_data['Sail Date'], errors='coerce', format='%m/%d/%y')
source_data.head()
```

```python
source_data['Sail Date'] = source_data['num_date'].dt.strftime("%A, %B %d, %Y")
source_data.head()
```

```python
source_data.to_csv('documents/celebrity-cruises.csv', index=False)
```
